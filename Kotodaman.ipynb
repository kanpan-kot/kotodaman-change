```
print('濁点、半濁点無しの五十音を入力してください。\n終了する場合は"f"を入力してください。')

while(1):
    BeforeChange= input()
    if BeforeChange == 'f' or BeforeChange == 'ｆ':
        print('終了しますか? [y/n]')
        flg = input()
        while(1):
            if flg == 'y' or flg == 'ｙ' or flg == 'n' or flg == 'ｎ':
                break
            else:
                print('"y"、または"n"を入力してください。')
                flg = input()
        if flg == 'y' or flg == 'ｙ':
            break
        elif flg == 'n' or flg == 'ｎ':
            continue
    for i in range(len(kawaridane)):
        if kawaridane[i] == BeforeChange:
            BeforeChange = gojuon[convert[i]]
    for i in range(len(gojuon)):
        if gojuon[i] == BeforeChange:
            print(str(change5) + '文字先:' + gojuon[(i+change5)%len(gojuon)])
            print(str(change6) + '文字先:' + gojuon[(i+change6)%len(gojuon)] + '\n')
            break
        else:
            if i == len(gojuon)-1:
                print('五十音を入力してください。\n')
                break
```
